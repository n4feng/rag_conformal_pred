```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cache_filename = (
    "data/result/Dragonball/dragonball_500_a=1.00_conformal_removal_cache.npy"
)
results = np.load(cache_filename, allow_pickle=True).item()

results["relavance"][0.05].keys()
```

```python
alphas = results["relavance"].keys()
confidence_methods = results.keys()

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10), constrained_layout=True)
axes = axes.flatten()

for idx, alpha in enumerate(alphas):
    data = [results[method][alpha]["fraction_removed"] for method in confidence_methods]
    axes[idx].boxplot(data, labels=confidence_methods)
    axes[idx].set_title(f" = {round(alpha, 2)}")
    axes[idx].set_ylabel("Removal Rate")
    axes[idx].tick_params(axis="x", rotation=45)
    # axes[idx].set_ylim(0.5, 1)  # Set the y-axis range for all subplots

# Hide any unused subplots
for idx in range(len(alphas), len(axes)):
    axes[idx].axis("off")

fig.suptitle(
    "Boxplot of removal rate for different scores at various  values", fontsize=16
)
plt.show()

fig.savefig(
    "data/result/Dragonball/dragonball_500_a=1.00_removal_rate_boxplot.png", dpi=800
)
```
